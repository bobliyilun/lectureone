Q13

Explain the naming rules, conventions and suggestions for:

a. constants

b. lists

c. functions

Here are the naming rules and conventions for each:

**a. Constants**  
- Use ALL_UPPERCASE letters with underscores between words.  
- Example: `MAX_SIZE`, `PI_VALUE`

**b. Lists**  
- Use lowercase letters with words separated by underscores.  
- Name should be plural to indicate multiple items.  
- Example: `students`, `user_names`, `scores_list`

**c. Functions**  
- Use lowercase letters with words separated by underscores (snake_case).  
- Name should describe the action or purpose.  
- Example: `calculate_total()`, `get_user_input()`, `print_report()`

Q14 Describe the following terms regarding Git version control:

a. Commit

b. Repository

c. Pull

d. Push

e. Branch


**a. Commit**  
A commit is a snapshot of changes made to files in a repository. It records what was changed and includes a message describing the changes.

**b. Repository**  
A repository is a storage location for your project files and their version history, managed by Git.

**c. Pull**  
Pull is the action of fetching and merging changes from a remote repository to your local repository.

**d. Push**  
Push is the action of sending your local commits to a remote repository.

**e. Branch**  
A branch is an independent line of development in a repository, allowing you to work on features or fixes separately from the main code.

Q15. Compare tuples and lists in Python, including what is the same and what is different between them. Describe the situations where you would choose one of these types over the other.

**Similarities:**
- Both tuples and lists can store multiple items of any type.
- Both are ordered collections and support indexing, slicing, and iteration.

**Differences:**
- **Lists** are mutable (can be changed: add, remove, or modify elements).
- **Tuples** are immutable (cannot be changed after creation).
- Lists use square brackets `[ ]`; tuples use parentheses `( )`.
- Lists have more built-in methods (like `append`, `remove`).

**When to use:**
- Use a **list** when you need to modify, add, or remove items.
- Use a **tuple** when you want a fixed collection of items that should not change, or for data that needs to be hashable (e.g., as dictionary keys).
